```python
import pandas as pd
#importing data

df=pd.read_csv('vending_machine_sales.csv')

df
```

```python
df1=df[(df.ID== "VJ300320609") & (df.Location=="GuttenPlans")]
p1=df1['Product'].value_counts(ascending=True)
placement_1=df1['RCoil'].value_counts()
placement_11=placement_1.head()
```
